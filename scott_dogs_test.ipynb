```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import path

dogs_intakes_df=pd.read_csv("Austin_Animal_Center_Intakes.csv")
dogs_intakes_df = dogs_intakes_df.loc[dogs_intakes_df['Animal Type'] == 'Dog']

print(dogs_intakes_df.shape)

dogs_intakes_df = dogs_intakes_df.sort_values(by=['MonthYear', 'Animal ID'])
print(dogs_intakes_df['MonthYear'].max())
print(dogs_intakes_df['MonthYear'].min())
print(dogs_intakes_df.shape)
print(dogs_intakes_df.head())
dogs_intakes_df.to_csv('dog_intakes.csv')
outcomes_df = pd.read_csv("Austin_Animal_Center_Outcomes.csv")

#Find only those dogs that resulted in adoption.
dogs_outcomes_df = outcomes_df.loc[(outcomes_df['Animal Type'] == 'Dog') & (outcomes_df['Outcome Type'] == 'Adoption')]
dogs_outcomes_df = dogs_outcomes_df.sort_values(by=['MonthYear', 'Animal ID'])


#Prints out the earliest date in the dogs_outcomes_df
print(pd.to_datetime(dogs_outcomes_df['MonthYear']).min())

#Prints the most recent date in the dogs_outcomes_df
print(dogs_outcomes_df['MonthYear'].max())

print(dogs_outcomes_df.head())

#Shows the number records in the dogs_outcomes_df
print(dogs_outcomes_df.shape)

#Puts the dogs_outcomes_df to a csv file.
dogs_outcomes_df.to_csv('dog_outcomes.csv')




```

```python
dog_df_merged = pd.merge(dogs_outcomes_df, dogs_intakes_df[['Animal ID', 
    'Name', 'DateTime', 'MonthYear', 'Found Location',
    'Intake Type', 'Intake Condition', 'Animal Type', 'Sex upon Intake',
    'Age upon Intake', 'Breed', 'Color']], 
    on='Animal ID', 
    how='left', 
    suffixes=('_outcome', '_intake'))


print(dog_df_merged.head())

print( dog_df_merged.mode()['Animal ID'][0])



#Normalize cuts off the time of the datetime object by converting it to midnight.
dog_df_merged['Date_outcome'] = pd.to_datetime(dog_df_merged['MonthYear_outcome']).dt.normalize()

dog_df_merged['Date_intake'] =  pd.to_datetime(dog_df_merged['MonthYear_intake']).dt.normalize()

dog_df_merged['stay_length']=abs((dog_df_merged['Date_outcome'])-\
    (dog_df_merged['Date_intake']))

dog_df_merged = dog_df_merged.sort_values(by=[ 'Animal ID', 'MonthYear_intake'])

print(dog_df_merged.shape)

dog_df_merged.head(30)
```
